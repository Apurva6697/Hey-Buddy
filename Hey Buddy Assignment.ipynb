```python
import speech_recognition as sr
```

```python
filename = "/content/machine-learning_speech-recognition_16-122828-0002.wav"
```

```python
# initialize the recognizer
r = sr.Recognizer()
```

```python
recognizer = sr.Recognizer()
# Load the audio file
with sr.AudioFile(filename) as source:
  # listen for the data (load audio to memory)
    audio_data = r.record(source)

    try:
        # Transcribe speech in English(# recognize (convert from speech to text))
        english_text = r.recognize_google(audio_data, language="en-US")
        print("English text:", english_text)

        # Transcribe speech in Spanish
        spanish_text = r.recognize_google(audio_data, language="es-ES")
        print("Spanish text:", spanish_text)

        # Add more languages as needed
        #french_text = recognizer.recognize_google(audio_data, language="fr-FR")
        #print("French text:", french_text)

    except sr.UnknownValueError:
        print("Sorry, I could not understand the audio")
    except sr.RequestError as e:
        print("Error occurred in connecting to the API. Check your internet connection:", e)
```
